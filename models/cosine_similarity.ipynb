```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
new_list = []
with open("../data/ingridients-food_.csv", "r", encoding="utf8") as f:
    # read lines
    lines = f.readlines()

# reomve first row
lines = lines[1:]

# for each line
for line in lines:
    # split the lines by comma
    line = line.lower()
    # .split(",")
    # remove the last element
    line = line[:-2]

    new_list.append(line)

with open("../data/ingridients-beauty_.csv", "r", encoding="utf8") as f:
    # read lines
    lines = f.readlines()

# reomve first row
lines = lines[1:]

# for each line
for line in lines:
    # split the lines by comma
    line = line.lower()
    # .split(",")
    # remove the last element
    line = line[:-2]

    new_list.append(line)


print(new_list)

# create a new dataframe
df = pd.DataFrame(new_list)
# give id
df['id'] = df.index
# add new columb df with just 1 st element separated by comma

df['title'] = df[0].str.split(',').str[0]



```

```python
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np
vec=TfidfVectorizer()
```

```python
vecs=vec.fit_transform(df[0].apply(lambda x: np.str_(x)))
```

```python
vecs.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim=cosine_similarity(vecs) 
```

```python
sim.shape
```

```python
sim
```

```python
sim[200][200]
```

```python
def recommend(product):
    idx=df[df["title"]==product]["id"].values[0]
    sim_scores=list(enumerate(sim[idx]))
    sim_scores=sorted(sim_scores,key=lambda x:x[1],reverse=True)
    sim_scores=sim_scores[1:11]
    food_idx=[i[0] for i in sim_scores]
    return df["title"].iloc[food_idx]
```

```python
input = "Patak's Aloo Tikki"

lst=recommend(input.lower())
lst
```
