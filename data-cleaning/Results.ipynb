```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate bar graph from results
# Result ={
#     'Model': 'Logistic Regression', 'Random Forest', 'Decision Tree',
# }

Accuracy ={
    'Logistic Regression':0.9467, 
    'Random Forest':0.8709677419354839, 
    'Decision Tree':0.9032258064516128,
}
F1 ={
    'Logistic Regression':0.97,  
    'Random Forest':0.87,   
    'Decision Tree':0.90,   
}
Precision ={
    'Logistic Regression':0.98, 
    'Random Forest':0.90,   
    'Decision Tree':0.92,   
}
Recall ={
    'Logistic Regression':0.97, 
    'Random Forest':0.86,   
    'Decision Tree':0.91,   
}

```

```python
titles = ['Accuracy', 'F1', 'Precision', 'Recall']
for indx, i in enumerate([Accuracy, F1, Precision, Recall]):
    plt.figure(figsize=(10, 5))
    plt.bar(i.keys(), i.values(), color=['#FAEBDD', '#E53940', '#5B1E51'])
    plt.xlabel("Model")
    plt.ylabel("Score")
    plt.title(titles[indx])
    plt.show()
```

```python
# export graphs to png
for indx, i in enumerate([Accuracy, F1, Precision, Recall]):
    plt.figure(figsize=(10, 5))
    plt.bar(i.keys(), i.values(), color=['#FAEBDD', '#E53940', '#5B1E51'])
    plt.xlabel("Model")
    plt.ylabel("Score")
    plt.title(titles[indx])
    plt.savefig(f'./graphs/{titles[indx]}.png')
```
