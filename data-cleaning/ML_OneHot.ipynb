# combining elements from food & beauty

```python
readFoodString = open('dataset/topFoodLabels.txt')
readBeautyString = open('dataset/topBeautyLabels.txt')
```

```python
# extract labels
foodLabels = readFoodString.read().split(",")
beautyLabels = readBeautyString.read().split(",") 
```

```python
# merge food and beauty labels and remove duplicates

labels = foodLabels + beautyLabels

indx = len(labels)
while(indx > 0):
    indx -= 1
    if labels.count(labels[indx]) > 1:
        print(indx)
        labels.pop(indx)
# for obj in labels:
#     if labels.count(obj) > 1:
#         # remove other duplicates except first
#         labels.remove(obj)


labels
```

```python
# export elements to file
topElementsFile = open("dataset/labels.txt", "w")
for element in labels:
    if(element == labels[-1]):
        topElementsFile.write(element)
    else:
        topElementsFile.write(element + ",")
topElementsFile.close()
```

# OneHot Encoding

```python
def onehotEncodeLabels(labels, inpString):
    # Split data by space and comma and to lower case
    inpString = inpString.lower()
    inpString = inpString.replace(" ", ",")
    inpString = inpString.split(",")
    # remove empty strings
    inpString = list(filter(None, inpString))
    # remove duplicates
    inpString = list(dict.fromkeys(inpString))
    # create one hot encoded array
    onehotEncoded = []
    for label in labels:
        if label in inpString:
            onehotEncoded.append(1)
        else:
            onehotEncoded.append(0)
    return onehotEncoded
```

```python
inputString = "Sugar, Milk Solids (23%), Cocoa Butter, Cocoa Solids, Emulsifiers (442, 476), Vegetable Fat, Maltitol, Flavoring Substances, Contains added flavors (Artificial (Vanilla and Chocolate)), Sweetener (Maltitol (965 I))."
onehotEncodeLabels(labels, inputString)
```
